### Filtrado por condiciones

```python
df[(df['favorites'] > 400) & (df['mentions'] > 20)] # CondiciÃ³n numÃ©rica
```

```python
df[df['full_text'].str.contains('Programming')]
```

```python
def calcular_ganancias(retweets):
    ganancias = retweets * random.randint(3, 5)
    return ganancias
    
df['ganancias'] = df['retweets'].apply(calcular_ganancias)

def popularidad(fila):
    resultado = fila['followees']/fila['followers']
    return resultado

df['popularidad'] = df.apply(popularidad, axis = 1)
df.head()
```

## AgrupaciÃ³n por datos categoricos

```python
df.groupby('country').mean() # AgrupaciÃ³n de una sola operaciÃ³n
```

```python
grupo = df.groupby('country').agg({ # con la funciÃ³n ".agg" puedo aplicar diferente funciones de agregaciÃ³n
    'followers':'sum',
    'mentions':'mean',
    'retweets':'max'
})

grupo[grupo['followers'] > 5000]
```

```python
import matplotlib.pyplot as plt

grupo['followers'].plot(kind = 'barh')
plt.show()
```
